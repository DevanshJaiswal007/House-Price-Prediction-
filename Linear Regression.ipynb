```python
from scipy.stats import iqr
q1=df['price'].quantile(0.20)
q3=df['price'].quantile(0.75)
inter_qr=iqr(df['price'])
print (q1)
print (q3)
print (inter_qr)
```

```python
df2=df[(df['price']>q1)& (df['price']<q3)]
df2.head()
```

```python
df2.shape
```

```python
box1=sns.boxplot(df2['price'])
```

```python
df_clean=df2.drop(columns=["street",'statezip',"country",'date','view','bathrooms','sqft_living','sqft_basement'])
df_clean.head(20)
```

```python
df_clean.shape
```

```python
df_clean=pd.get_dummies(df_clean)
df_clean.head(10)
```

```python
x= df_clean.drop(columns='price')
y=df_clean['price']

print(x.shape)
print(y.shape)
```

```python
from sklearn.model_selection import train_test_split

xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2)
print(xtrain.shape)
print(xtest.shape)
print(ytrain.shape)
print(ytest.shape)
```

```python
import statsmodels.api as sm

model=sm.OLS(ytrain,xtrain).fit()
```

```python
print(model.summary())
```

```python
pred=model.predict(xtest)
data=list(zip(ytest,pred))
```

```python
comptab=pd.DataFrame(data,columns=['Actual','Predicted'])
comptab.head(20)
```

```python
from sklearn.metrics import mean_squared_error
from math import sqrt

rmse=sqrt(mean_squared_error(ytest,pred))

print (rmse)
```
